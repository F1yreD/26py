```python
import numpy as np
import matplotlib.pyplot as plt

# 设定一元二次方程：f(x) = (x-3)^2 + 4
# 我们的目标是最小化 f(x)

def f(x):
    return (x - 3) ** 2 + 4

def grad_f(x):
    # f(x) = (x-3)^2 + 4，导数为2*(x-3)
    return 2 * (x - 3)

# 初始化
x = 0.0
learning_rate = 0.1
num_steps = 25

x_history = [x]
f_history = [f(x)]

# 梯度下降过程
for step in range(num_steps):
    grad = grad_f(x)
    x = x - learning_rate * grad
    x_history.append(x)
    f_history.append(f(x))

# 可视化
plt.figure(figsize=(8, 5))

# 绘制曲线
x_plot = np.linspace(-1, 7, 100)
y_plot = f(x_plot)
plt.plot(x_plot, y_plot, label="f(x) = (x-3)² + 4")

# 梯度下降路径
plt.scatter(x_history, f_history, color='red', zorder=5)
plt.plot(x_history, f_history, color='red', linestyle='--', label='Gradient Descent')

plt.title("Gradient Descent on f(x) = (x-3)² + 4")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid()
plt.show()

print("最终收敛的x值:", x)
print("最终的f(x):", f(x))
print("x_history:", x_history)
print("f_history:", f_history)
```
